```python
import pandas as pd
import numpy as np
import seaborn as sn
#---------------------- RDKit packages
from rdkit.Chem import AllChem
from rdkit.Chem import rdMolDescriptors
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
from rdkit import DataStructs
from rdkit.ML.Cluster import Butina
#------------------- progress bar
from tqdm import tqdm
#------------------- hide warning
import warnings
warnings.filterwarnings('ignore')
dataset = pd.read_csv('/content/sample_data/dbfda-ligands.csv')
dataset.head(10)
```

```python
# Include 2D structure of molecules
# https://github.com/rdkit/rdkit/blob/master/rdkit/Chem/PandasTools.py

from rdkit.Chem import PandasTools
PandasTools.AddMoleculeColumnToFrame(dataset,'smiles', 'Structure')
dataset.head()
```

```python
from rdkit.Chem import Draw
Draw.MolsToGridImage([x for x in dataset["Structure"]][:98], maxMols=100,
                     legends=[str(x) for x in dataset.zinc_id], subImgSize=(200,200), molsPerRow=7)
```
